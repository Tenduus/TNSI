# Grand Oral : 

### Sujet : *Les algorithmes de cryptographie symétrique sont-ils encore pertinents dans la sécuristation des données en 2024 ?*

### Sujets à aborder (plan) : 

En 2024, les algorithmes de cryptographie symétrique restent extrêmement pertinents pour la sécurisation des données pour plusieurs raisons :

#### 1. **Efficacité et Rapidité**

- **Performance** : Les algorithmes de cryptographie symétrique sont généralement plus rapides et nécessitent moins de ressources computationnelles que les algorithmes asymétriques. Ils sont donc idéaux pour le chiffrement de grandes quantités de données.
- **Utilisation Pratique** : Leur efficacité les rend adaptés aux applications où la rapidité et la faible latence sont cruciales, telles que les communications en temps réel, le chiffrement de disque et les réseaux VPN.

#### 2. **Sécurité**

- **Robustesse** : Les algorithmes modernes de cryptographie symétrique, tels que AES (Advanced Encryption Standard), sont extrêmement sécurisés lorsqu'ils sont correctement implémentés et utilisés avec des clés de longueur suffisante (par exemple, 256 bits).
- **Longévité** : Bien qu'il y ait des avancées en informatique quantique, la cryptographie symétrique est considérée comme moins vulnérable à court terme. Les algorithmes symétriques nécessitent de nouvelles techniques quantiques pour casser, et ces techniques ne sont pas aussi avancées que celles pour les algorithmes asymétriques.

#### 3. **Facilité de Mise en Œuvre**

- **Simplicité** : Comparés aux algorithmes asymétriques, les algorithmes symétriques sont souvent plus simples à implémenter et à comprendre, ce qui réduit le risque d'erreurs de mise en œuvre.
- **Interopérabilité** : Ils sont largement supportés par divers protocoles et standards, facilitant leur intégration dans de nombreux systèmes et applications.

#### 4. **Cas d'Utilisation Spécifiques**

- **Chiffrement de Données en Vol** : Lors de la transmission de données, comme dans les protocoles TLS (Transport Layer Security), la cryptographie symétrique est utilisée pour le chiffrement des sessions après un échange initial de clés via des algorithmes asymétriques.
- **Chiffrement de Données au Repos** : Pour protéger les données stockées (sur disque, dans des bases de données), les algorithmes symétriques sont souvent utilisés pour leur rapidité et efficacité.

#### 5. **Résilience Quantique à Court Terme**

- **Problèmes de Capacité** : Les ordinateurs quantiques capables de casser les algorithmes de cryptographie symétrique comme AES nécessiteraient un nombre de qubits beaucoup plus important et une correction d'erreurs significative. Cela est encore considéré comme un défi à long terme.

#### 6. **Combinés avec des Algorithmes Asymétriques**

- **Sécurisation Hybride** : Souvent, les systèmes sécurisés combinent des algorithmes symétriques et asymétriques pour tirer parti des avantages de chacun. Par exemple, l'échange de clés initial utilise des algorithmes asymétriques, puis les algorithmes symétriques prennent le relais pour le chiffrement des données.

#### Conclusion

En 2024, les algorithmes de cryptographie symétrique restent essentiels en raison de leur efficacité, leur robustesse, leur facilité de mise en œuvre et leur pertinence pour un large éventail d'applications. Bien que la menace quantique soit une considération pour l'avenir, les algorithmes symétriques restent une partie intégrante des stratégies de sécurisation des données modernes.

### Chiffrement symétrique :

Le chiffrement symétrique trouve ses origines dans l'histoire ancienne de la cryptographie. Des méthodes de chiffrement symétriques rudimentaires étaient utilisées dès l'Antiquité pour protéger les communications militaires et diplomatiques. Un exemple célèbre est le chiffrement de César, où chaque lettre du message était décalée de 3 rang.
Dans le chiffrement symétrique, il y a une seule clé de déchiffrement (qui est identique) et qui va permettre de déchiffrer le message.

![Chiffrement symétrique](https://cellar-c2.services.clever-cloud.com/s3.primx.eu/uploads/2022/04/PRIMX_symetric_encryption-1024x377.png)

Evidemment, cela pose des problèmes de sécurité liés au partage de cette clé (Chiffre de Vernam). Le risque étant quelle soit visible si cette échange ce fait via internet. Il est donc préférable de la remettre en main propre afin de limiter ce risque, ce qui, je vous l'accorde est très peu pratique surtout si notre interlocuteur se trouve à l'autre bout du monde.

### chiffrement asymétrique :

Le concept de chiffrement asymétrique, également connu sous le nom de cryptographie à clé publique, a été développé dans les années 1970 par Whitfield Diffie, Martin Hellman et Ralph Merkle. Leur travail novateur a été publié en 1976 dans un article intitulé "New Directions in Cryptography"

Le chiffrement asymétrique est caractérisé par l'utilisation de **clés publique**/**clé privé**.

![Chiffrement asymétrique](https://www.tice-education.fr/images/stories/img/chiffrement-asymetrique.JPG)

Dans cette exemple, si l'utilisateur *A* veut envoyer "un message" à l'utilisateur *B*, il doit lui demander sa **clé publique**, qui est visible par tous. Cette clé sert uniquement à **chiffré le fichier**. Ainsi une fois le message chiffré, l'utilisateur *A* peut envoyé le fichier en toute sécurité à l'utilisateur *B* qui pourra le déchiffrer grâce à sa **clé privé** qu'il est le seul à posséder (d'où son nom). 




### Les ordinateurs quantique : 

#### Définition : 
Un ordinateur quantique, calculateur quantique, processeur quantique ou système informatique quantique, utilise les propriétés quantiques de la matière, telles que la superposition et l'intrication, afin d'effectuer des opérations sur des données.

#### Propriétés : 

**Intrication :** Si deux particules se rencontrent, elles vont garder un lien étroit même à l'autre bout de l'univers. Si l'une d'elle change d'état, l'autre changera aussi.

- Amélioration des communications car elles ne transitent pas par des éléments physique. 
- Communication instantané.

**Superposition :**

Des particules se trouve en même temps dans toutes ses configurations possibles jusqu'a ce que l'on mesure létat de cette particule.

**Interférence :**

*- Interférence Constructive :*

Lorsque deux ondes de probabilité sont en phase (les crêtes et les creux coïncident), elles se renforcent mutuellement.
La probabilité résultante est plus élevée dans certaines régions.

*- Interférence Destructive :*

Lorsque deux ondes de probabilité sont en opposition de phase (les crêtes d'une onde coïncident avec les creux de l'autre), elles se neutralisent.
La probabilité résultante est réduite ou annulée dans certaines régions.

En 2024, les algorithmes de cryptographie symétrique restent extrêmement pertinents pour la sécurisation des données pour plusieurs raisons :

#### 1. **Efficacité et Rapidité**

- **Performance** : Les algorithmes de cryptographie symétrique sont généralement plus rapides et nécessitent moins de ressources computationnelles que les algorithmes asymétriques. Ils sont donc idéaux pour le chiffrement de grandes quantités de données.
- **Utilisation Pratique** : Leur efficacité les rend adaptés aux applications où la rapidité et la faible latence sont cruciales, telles que les communications en temps réel, le chiffrement de disque et les réseaux VPN.

#### 2. **Sécurité**

- **Robustesse** : Les algorithmes modernes de cryptographie symétrique, tels que AES (Advanced Encryption Standard), sont extrêmement sécurisés lorsqu'ils sont correctement implémentés et utilisés avec des clés de longueur suffisante (par exemple, 256 bits).
- **Longévité** : Bien qu'il y ait des avancées en informatique quantique, la cryptographie symétrique est considérée comme moins vulnérable à court terme. Les algorithmes symétriques nécessitent de nouvelles techniques quantiques pour casser, et ces techniques ne sont pas aussi avancées que celles pour les algorithmes asymétriques.

#### 3. **Facilité de Mise en Œuvre**

- **Simplicité** : Comparés aux algorithmes asymétriques, les algorithmes symétriques sont souvent plus simples à implémenter et à comprendre, ce qui réduit le risque d'erreurs de mise en œuvre.
- **Interopérabilité** : Ils sont largement supportés par divers protocoles et standards, facilitant leur intégration dans de nombreux systèmes et applications.

#### 4. **Cas d'Utilisation Spécifiques**

- **Chiffrement de Données en Vol** : Lors de la transmission de données, comme dans les protocoles TLS (Transport Layer Security), la cryptographie symétrique est utilisée pour le chiffrement des sessions après un échange initial de clés via des algorithmes asymétriques.
- **Chiffrement de Données au Repos** : Pour protéger les données stockées (sur disque, dans des bases de données), les algorithmes symétriques sont souvent utilisés pour leur rapidité et efficacité.

#### 5. **Résilience Quantique à Court Terme**

- **Problèmes de Capacité** : Les ordinateurs quantiques capables de casser les algorithmes de cryptographie symétrique comme AES nécessiteraient un nombre de qubits beaucoup plus important et une correction d'erreurs significative. Cela est encore considéré comme un défi à long terme.

#### 6. **Combinés avec des Algorithmes Asymétriques**

- **Sécurisation Hybride** : Souvent, les systèmes sécurisés combinent des algorithmes symétriques et asymétriques pour tirer parti des avantages de chacun. Par exemple, l'échange de clés initial utilise des algorithmes asymétriques, puis les algorithmes symétriques prennent le relais pour le chiffrement des données.

#### Conclusion

En 2024, les algorithmes de cryptographie symétrique restent essentiels en raison de leur efficacité, leur robustesse, leur facilité de mise en œuvre et leur pertinence pour un large éventail d'applications. Bien que la menace quantique soit une considération pour l'avenir, les algorithmes symétriques restent une partie intégrante des stratégies de sécurisation des données modernes.

### Chiffrement symétrique :

Le chiffrement symétrique trouve ses origines dans l'histoire ancienne de la cryptographie. Des méthodes de chiffrement symétriques rudimentaires étaient utilisées dès l'Antiquité pour protéger les communications militaires et diplomatiques. ### chiffrement asymétrique :

Le concept de chiffrement asymétrique, également connu sous le nom de cryptographie à clé publique, a été développé dans les années 1970 par Whitfield Diffie, Martin Hellman et Ralph Merkle. Leur travail novateur a été publié en 1976 dans un article intitulé "New Directions in Cryptography"

Le chiffrement asymétrique est caractérisé par l'utilisation de **clés publique**/**clé privé**.

![Chiffrement asymétrique](https://www.tice-education.fr/images/stories/img/chiffrement-asymetrique.JPG)

Dans cette exemple, si l'utilisateur *A* veut envoyer "un message" à l'utilisateur *B*, il doit lui demander sa **clé publique**, qui est visible par tous. Cette clé sert uniquement à **chiffré le fichier**. Ainsi une fois le message chiffré, l'utilisateur *A* peut envoyé le fichier en toute sécurité à l'utilisateur *B* qui pourra le déchiffrer grâce à sa **clé privé** qu'il est le seul à posséder (d'où son nom). 




### Exemple de chiffrement symétrique : 

#### le Chiffre de Vernam : 

Le **masque jetable**, également appelé **chiffre de Vernam**, est un algorithme de cryptographie inventé par Gilbert Vernam en **1917**
 et perfectionné par Joseph Mauborgne, qui rajouta la notion de clé aléatoire.

Le chiffrement par la méthode du masque jetable consiste à combiner le message en clair avec une clé présentant les caractéristiques très particulières suivantes :

- La clé doit être une suite de caractères au moins aussi longue que le message à chiffrer.
- Les caractères composant la clé doivent être choisis de façon totalement aléatoire.
- Chaque clé, ou « masque », ne doit être utilisée qu'une seule fois (d'où le nom de masque jetable).

##### Fonctionnement :

1. **Génération de la clé** : Pour commencer, une clé secrète de la même longueur que le message à chiffrer est générée. Cette clé est composée de bits aléatoires ou pseudo-aléatoires.

2. **Chiffrement** : Le message original est représenté sous forme binaire, de même que la clé. Pour chiffrer le message, chaque bit du message original est combiné avec le bit correspondant de la clé en utilisant l'opération de XOR. Si les deux bits sont identiques, le résultat est 0 ; s'ils sont différents, le résultat est 1. Le message chiffré est ensuite transmis.

3. **Déchiffrement** : Pour déchiffrer le message, le même processus est appliqué. Chaque bit du message chiffré est combiné avec le bit correspondant de la clé en utilisant l'opération de XOR. Comme l'opération de XOR est son propre inverse (A XOR B XOR B = A), le résultat est le message original.

L'opération de XOR est choisie pour plusieurs raisons dans le chiffrement de Vernam :

- **Simplicité** : L'opération de XOR est simple à calculer et rapide à exécuter, ce qui en fait un choix efficace pour le chiffrement.

- **Réversibilité** : Comme mentionné, l'opération de XOR est son propre inverse, ce qui facilite le processus de déchiffrement.

- **Propriétés mathématiques** : L'opération de XOR possède des propriétés intéressantes qui la rendent utile en cryptographie, notamment sa linéarité (f(x XOR y) = f(x) XOR f(y)) et sa non-linéarité, en fonction du contexte.

Le chiffrement de Vernam est considéré comme inviolable s'il est correctement implémenté avec une clé aléatoire ou pseudo-aléatoire de la même longueur que le message et utilisée une seule fois (OTP - One-Time Pad). Cependant, la gestion de clés pour OTP reste un défi important dans la pratique.

### **protocole TLS (Transport Layer Security)**

### Comment TLS Utilise les Deux Types de Cryptographie

1. **Établissement de la Connexion** :
   - **Échange de Clés** : TLS utilise la cryptographie asymétrique pour échanger une clé de session de manière sécurisée. Typiquement, RSA ou ECC est utilisé pour chiffrer et transmettre cette clé.
   - **Authentification** : Le serveur (et parfois le client) utilise des certificats numériques basés sur RSA ou ECC pour prouver son identité.

2. **Chiffrement de la Session** :
   - **Chiffrement Symétrique** : Une fois la clé de session échangée, TLS utilise des algorithmes de chiffrement symétrique comme AES ou ChaCha20 pour chiffrer les données échangées durant la session. Cette clé de session symétrique est utilisée pour garantir des communications rapides et sécurisées.

### Avantages de Combiner les Deux Types

1. **Sécurité et Performance** :
   - **Cryptographie Asymétrique** : Assure un échange de clés sécurisé et l'authentification. Cela résout le problème de la distribution sécurisée des clés symétriques.
   - **Cryptographie Symétrique** : Offre une performance supérieure pour le chiffrement de grandes quantités de données une fois la clé de session échangée.

2. **Robustesse Contre les Attaques** :
   - La combinaison permet d'utiliser les forces de chaque type de cryptographie tout en minimisant leurs faiblesses.
   - L'usage asymétrique pour l'échange de clés empêche les attaquants d'intercepter et de comprendre la clé de session, même s'ils interceptent la communication.

### Autres Exemples

1. **PGP (Pretty Good Privacy)** :
   - Utilisé pour le chiffrement des courriels.
   - Utilise la cryptographie asymétrique pour chiffrer une clé de session symétrique, qui est ensuite utilisée pour chiffrer le message réel.

2. **S/MIME (Secure/Multipurpose Internet Mail Extensions)** :
   - Utilisé pour sécuriser les courriels.
   - Fonctionne de manière similaire à PGP en utilisant la cryptographie asymétrique pour l'échange sécurisé de clés symétriques.

Les protocoles comme TLS, PGP et S/MIME exploitent cette combinaison pour assurer des communications sécurisées sur Internet, garantissant à la fois la confidentialité et l'intégrité des données.

### Les ordinateurs quantique : 

#### Définition : 
Un ordinateur quantique, calculateur quantique, processeur quantique ou système informatique quantique, utilise les propriétés quantiques de la matière, telles que la superposition et l'intrication, afin d'effectuer des opérations sur des données.

#### Propriétés : 

**Intrication :** Si deux particules se rencontrent, elles vont garder un lien étroit même à l'autre bout de l'univers. Si l'une d'elle change d'état, l'autre changera aussi.

- Amélioration des communications car elles ne transitent pas par des éléments physique. 
- Communication instantané.

**Superposition :**

Des particules se trouve en même temps dans toutes ses configurations possibles jusqu'a ce que l'on mesure létat de cette particule.

**Interférence :**

*- Interférence Constructive :*

Lorsque deux ondes de probabilité sont en phase (les crêtes et les creux coïncident), elles se renforcent mutuellement.
La probabilité résultante est plus élevée dans certaines régions.

*- Interférence Destructive :*

Lorsque deux ondes de probabilité sont en opposition de phase (les crêtes d'une onde coïncident avec les creux de l'autre), elles se neutralisent.
La probabilité résultante est réduite ou annulée dans certaines régions.

**Algorithme de Shor** : 
 
 L'algorithme de Shor est un algorithme quantique qui permet de factoriser des grands nombres entiers de manière extrêmement efficace, bien plus rapidement que les meilleurs algorithmes classiques connus. Voici une explication simplifiée de son fonctionnement. L'algorithme de Shor est un algorithme quantique capable de factoriser des grands nombres entiers en temps polynomial, c'est-à-dire beaucoup plus rapidement que les meilleurs algorithmes classiques.

##### Principe de Sécurité de RSA

1. **Clé Publique et Privée** :
   - RSA utilise une paire de clés : une clé publique pour chiffrer les messages et une clé privée pour les déchiffrer.
   - La clé publique est composée de deux nombres : \( n \) (un grand nombre entier) et \( e \).
   - La clé privée est dérivée de la clé publique et repose sur deux grands nombres premiers, \( p \) et \( q \), qui multiplient pour donner \( n \).

2. **Difficulté de la Factorisation** :
   - La sécurité de RSA repose sur la difficulté de factoriser \( n \) en ses deux facteurs premiers \( p \) et \( q \).
   - Connaître \( p \) et \( q \) permet de dériver la clé privée, mais factoriser \( n \) est très difficile pour les grands nombres avec les méthodes classiques.

##### Comment l'Algorithme de Shor Casse RSA

1. **Algorithme de Shor** :
   - L'algorithme de Shor est un algorithme quantique capable de factoriser des grands nombres entiers en temps polynomial, c'est-à-dire beaucoup plus rapidement que les meilleurs algorithmes classiques.

2. **Étape Clé : Trouver les Facteurs \( p \) et \( q \)** :
   - L'algorithme de Shor prend comme entrée le nombre \( n \) de la clé publique RSA.
   - Il utilise des propriétés quantiques comme la superposition et l'interférence pour trouver rapidement les facteurs premiers \( p \) et \( q \).

3. **Accès à la Clé Privée** :
   - Une fois que \( p \) et \( q \) sont connus, il est facile de calculer la clé privée.
   - Cela permet à une personne malveillante de déchiffrer les messages chiffrés avec la clé publique sans autorisation.

##### Simplifié

- **RSA** : Protège les données en utilisant un grand nombre \( n \) difficile à factoriser.
- **Algorithme de Shor** : Un outil puissant pour un ordinateur quantique qui peut rapidement factoriser ce grand nombre \( n \) et découvrir les nombres premiers \( p \) et \( q \).
- **Conséquence** : Une fois \( p \) et \( q \) connus grâce à l'algorithme de Shor, le chiffrement RSA est cassé car la clé privée peut être calculée et utilisée pour déchiffrer les données protégées.

En résumé, l'algorithme de Shor menace RSA parce qu'il peut factoriser rapidement le grand nombre \( n \) en ses facteurs premiers \( p \) et \( q \), ce qui permet de trouver la clé privée et de casser le chiffrement RSA.
